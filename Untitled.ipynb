## Regrid

```python
min_lat = ds_agcd.lat_bnds.values.min()
max_lat = ds_agcd.lat_bnds.values.max()
min_lon = ds_agcd.lon_bnds.values.min()
max_lon = ds_agcd.lon_bnds.values.max()
print(min_lat, max_lat, min_lon, max_lon)
```

```python
original_grid = {'lon': ds_agcd['lon'].values,
                 'lat': ds_agcd['lat'].values,
                 'lon_b': np.append(ds_agcd['lon_bnds'].values[:, 0], ds_agcd['lon_bnds'].values[-1, 1]),
                 'lat_b': np.append(ds_agcd['lat_bnds'].values[:, 0], ds_agcd['lat_bnds'].values[-1, 1])}
```

```python
original_grid
```

```python
ds_cafe = xr.open_zarr('/g/data/xv83/dcfp/CAFE-f6/c5-d60-pX-f6-19811101/atmos_isobaric_daily.zarr.zip')
ds_cafe
```

```python
min_latb_index = np.searchsorted(ds_cafe['latb'].values, min_lat)
max_latb_index = np.searchsorted(ds_cafe['latb'].values, max_lat)
min_lonb_index = np.searchsorted(ds_cafe['lonb'].values, min_lon)
max_lonb_index = np.searchsorted(ds_cafe['lonb'].values, max_lon)

new_grid = {'lon': ds_cafe['lon'].values[min_lonb_index : max_lonb_index - 1],
            'lat': ds_cafe['lat'].values[min_latb_index : max_latb_index - 1],
            'lon_b': ds_cafe['lonb'].values[min_lonb_index : max_lonb_index],
            'lat_b': ds_cafe['latb'].values[min_latb_index : max_latb_index]}
```

```python
new_grid
```

```python
regridder = xe.Regridder(original_grid, new_grid, 'conservative')
regridder
```

```python
ds_out = regridder(ds_agcd[['precip']])
```

```python
lat_bnds = np.column_stack([new_grid['lat_b'][0:-1], new_grid['lat_b'][1:]])
lon_bnds = np.column_stack([new_grid['lon_b'][0:-1], new_grid['lon_b'][1:]])
```

```python
ds_out = ds_out.assign_coords({'lat_bnds': (('lat', 'nv'), lat_bnds),
                               'lon_bnds': (('lon', 'nv'), lon_bnds),
                               'time_bnds': ds_agcd['time_bnds']})
```

```python
ds_out
```

```python
attrs = ds_agcd.attrs
for att in ['geospatial_lat_min', 'geospatial_lat_max', 'geospatial_lon_min', 'geospatial_lon_max']:
    del attrs[att]
```

```python
ds_out.attrs = attrs
```

```python
ds_out
```

```python
fig = plt.figure(figsize=(12, 7))

ax = fig.add_subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()

ds_out['precip'].isel(time=0).plot(ax=ax, transform=ccrs.PlateCarree())

ax.plot([min_lon, max_lon], [min_lat, min_lat], color='red')
ax.plot([min_lon, max_lon], [max_lat, max_lat], color='red')
ax.plot([min_lon, min_lon], [min_lat, max_lat], color='red')
ax.plot([max_lon, max_lon], [min_lat, max_lat], color='red')

ax.set_extent([min_lon - 5, max_lon + 5, min_lat - 5, max_lat + 5], crs=ccrs.PlateCarree())

plt.show()
```
