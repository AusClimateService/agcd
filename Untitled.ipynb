## Regrid

```python
min_lat = ds_agcd.lat_bnds.values.min()
max_lat = ds_agcd.lat_bnds.values.max()
min_lon = ds_agcd.lon_bnds.values.min()
max_lon = ds_agcd.lon_bnds.values.max()
print(min_lat, max_lat, min_lon, max_lon)
```

```python
agcd_grid = {'lon': ds_agcd['lon'].values,
             'lat': ds_agcd['lat'].values,
             'lon_b': np.arange(min_lon, max_lon + 0.025, 0.05).round(3),
             'lat_b': np.arange(min_lat, max_lat + 0.025, 0.05).round(3)}
```

```python
agcd_grid
```

```python
ds_cafe = xr.open_zarr('/g/data/xv83/dcfp/CAFE-f6/c5-d60-pX-f6-19811101/atmos_isobaric_daily.zarr.zip')
ds_cafe
```

```python
ds_cafe = ds_cafe.sel(lat=slice(min_lat, max_lat), lon=slice(min_lon, max_lon))
```

```python
cafe_grid = {'lon': ds_cafe['lon'].values,
             'lat': ds_cafe['lat'].values,
             'lon_b': ds_cafe['lonb'].values[45:64],
             'lat_b': ds_cafe['latb'].values[23:41]}
```

```python
cafe_grid
```

```python
regridder1 = xe.Regridder(agcd_grid, cafe_grid, 'conservative')
regridder1
```

```python
ds_out1 = regridder1(ds_agcd[['precip']])
```

```python
ds_out1.compute()
```

```python
fig = plt.figure(figsize=(12, 7))

ax = fig.add_subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()

ds_out1['precip'].isel(time=0).plot(ax=ax, transform=ccrs.PlateCarree())

ax.plot([min_lon, max_lon], [min_lat, min_lat], color='red')
ax.plot([min_lon, max_lon], [max_lat, max_lat], color='red')
ax.plot([min_lon, min_lon], [min_lat, max_lat], color='red')
ax.plot([max_lon, max_lon], [min_lat, max_lat], color='red')

ax.set_extent([min_lon - 5, max_lon + 5, min_lat - 5, max_lat + 5], crs=ccrs.PlateCarree())

plt.show()
```
