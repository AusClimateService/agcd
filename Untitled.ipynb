```python
import glob

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
import cmdline_provenance as cmdprov

import zarr_utils
```

```python
infiles = glob.glob('/g/data/zv2/agcd/v2/precip/total/r005/01month/agcd_v2_precip_total_r005_monthly_*.nc')
infiles.sort()
infiles
```

```python
ds = xr.open_mfdataset(infiles, concat_dim='time', combine='nested', data_vars='minimal', coords='minimal', compat='override')
ds
```

```python
ds.attrs['history'] = cmdprov.new_log()
ds.attrs['history']
```

```python
outfile = '/g/data/xv83/dbi599/agcd/agcd_v2_precip_total_r005_monthly_1900-2020.zarr.zip'
zarr_utils.to_zarr(ds, outfile)
```

```python
ds_ref = xr.open_zarr('/g/data/xv83/dcfp/CAFE-f6/c5-d60-pX-f6-19811101/atmos_isobaric_month.zarr.zip')
ds_ref
```

```python
ds_ref_grid = ds_ref[['lat', 'lon']]
```

```python
ds_ref_grid
```

```python
ds_ref_grid = ds_ref_grid.drop(['area', 'zsurf'])
```

```python
regridder = xe.Regridder(ds, ds_ref_grid, 'conservative')
regridder 
```

```python
ds_out = regridder(ds)
#regridder.clean_weight_file()
```

```python
ds_out
```

```python
min_lat = ds.lat_bnds.values.min()
max_lat = ds.lat_bnds.values.max()
min_lon = ds.lon_bnds.values.min()
max_lon = ds.lon_bnds.values.max()
print(min_lat, max_lat, min_lon, max_lon)
```

```python
fig = plt.figure(figsize=(12, 7))

ax = fig.add_subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()

ds_out['precip'].isel(time=0).plot(ax=ax, transform=ccrs.PlateCarree())

ax.plot([min_lon, max_lon], [min_lat, min_lat], color='red')
ax.plot([min_lon, min_lon], [min_lat, max_lat], color='red')

ax.set_extent([min_lon - 5, max_lon + 5, min_lat - 5, max_lat + 5], crs=ccrs.PlateCarree())

plt.show()
```
